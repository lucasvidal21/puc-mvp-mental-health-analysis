## 3.2 - Normalização

A normalização consiste em ajustar os valores das variáveis para uma mesma escala, geralmente entre 0 e 1, sendo útil principalmente em modelos de aprendizado de máquina.

No entanto, neste trabalho, não foi necessária a aplicação de normalização, uma vez que as variáveis numéricas já se encontram na mesma escala, representando proporções da população.

Dessa forma, a manutenção dos valores originais preserva a interpretação dos dados, sendo mais adequada para a análise exploratória realizada.


## 3.3 - Padronização

A padronização, também conhecida como Z-score scaling, consiste em transformar os dados para que apresentem média igual a zero e desvio padrão igual a um, sendo útil principalmente em modelos de aprendizado de máquina sensíveis à escala das variáveis.

No entanto, neste trabalho, não foi necessária a aplicação da padronização, uma vez que as variáveis numéricas já se encontram em uma mesma escala, representando proporções da população.

Além disso, como o foco da análise é exploratório, a manutenção dos valores originais permite uma interpretação mais direta dos resultados.
